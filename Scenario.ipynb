```python
# VOTING ELIGIBILITY

age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
#Find the Largest Number in a List

numbers = [10, 25, 5, 40, 15]

largest = max(numbers)
print("Largest number is:", largest)
```

```python
#Salary Bonus Calculation

salary = float(input("Enter salary: "))

if salary > 50000:
    bonus = salary * 0.10
else:
    bonus = 0

print("Bonus amount:", bonus)
```

```python
#Even or Odd Number

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
#Reverse a Word or Sentence

text = input("Enter text: ")

reversed_text = text[::-1]
print("Reversed text:", reversed_text)
```

```python
#Pass or Fail (Minimum 40 Marks)

marks = int(input("Enter marks: "))

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
#Retail Discount Calculation

total_amount = float(input("Enter total order amount: "))

if total_amount > 100:
    discount = total_amount * 0.20
else:
    discount = 0

final_amount = total_amount - discount
print("Final amount to pay:", final_amount)
```

```python
#Bank Withdrawal Check

balance = 5000
withdraw_amount = float(input("Enter withdrawal amount: "))

if withdraw_amount <= balance:
    balance -= withdraw_amount
    print("Withdrawal successful")
    print("Remaining balance:", balance)
else:
    print("Insufficient balance")
```

```python
#Leap Year Check

year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

```python
#Extract Even Numbers from a List

numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers = []

for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)

print("Even numbers:", even_numbers)
```
