```python
# print 0 to 20 by using range
for i in range(20):
    print(i)
```

```python
# print range 10 to 20
for i in range(10,20):
    print(i,end=' ')
```

```python
# Print number of items in the list by using 'len'
lst=[10,20,14,55,43,87,76]
print(lst,'\n' 'Number of item in the List:',len(lst))
```
