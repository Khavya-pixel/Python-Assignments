```python
#Prompt until a valid integer
while True:
    user_input = input("Enter a valid integer: ")
    try:
        num = int(user_input)
        print("Valid integer entered:", num)
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

```python
#Most frequently occurring number in a list
from collections import Counter

numbers = [1, 2, 2, 3, 4, 2, 3]
freq = Counter(numbers)
most_common = freq.most_common(1)[0][0]
print("Most frequent number:", most_common)
```

```python
#Check if two strings are anagrams
def are_anagrams(str1, str2):
    return sorted(str1) == sorted(str2)

print(are_anagrams("listen", "silent"))  
print(are_anagrams("hello", "world"))    
```

```python
#Reverse the order of words in a sentence
sentence = "Python is fun"
reversed_sentence = " ".join(sentence.split()[::-1])
print(reversed_sentence) 
```

```python
#Find the missing number in a sequence
numbers = [1, 2, 4, 5]  
n = len(numbers) + 1
missing_number = n*(n+1)//2 - sum(numbers)
print("Missing number:", missing_number)
```

```python
#ATM withdrawal check
balance = 5000
withdraw = 6000

if withdraw <= balance:
    balance -= withdraw
    print("Withdrawal successful. Remaining balance:", balance)
else:
    print("Insufficient balance!")
```

```python
#Check for duplicates in a list
numbers = [1, 2, 3, 2, 4]
has_duplicates = len(numbers) != len(set(numbers))
print("Contains duplicates:", has_duplicates)
```

```python
#Sum of digits of a number
num = 1234
digit_sum = sum(int(d) for d in str(num))
print("Sum of digits:", digit_sum)  
```

```python
#Check if a sentence is a pangram
import string

def is_pangram(sentence):
    alphabet = set(string.ascii_lowercase)
    return alphabet <= set(sentence.lower())

sentence = "The quick brown fox jumps over a lazy dog"
print("Is pangram:", is_pangram(sentence)) 
```
